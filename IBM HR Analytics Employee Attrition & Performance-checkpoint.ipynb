# detect columns where there is only 1 unique value

```python
single_unique_cols = [col for col in df.columns if df[col].nunique() == 1]
df=df.drop(columns=single_unique_cols)
```

```python
df=df.drop('EmployeeNumber', axis=1)
```

```python
gui=show(df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

#identifying categorical columns
cat_cols = df.select_dtypes(include='object').columns
```

```python
le = LabelEncoder()
for col in cat_cols:
    n_unique = df[col].nunique()
    
    if n_unique == 2:
       
        df[col] = le.fit_transform(df[col])
        print(f"Label encoded: {col}")
        
    elif n_unique <= 10:
        
        df = pd.get_dummies(df, columns=[col], drop_first=True)
        print(f"One-hot encoded: {col}")
    else:
        # Too many unique values
        print(col)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Attrition')
plt.title("Class Balance: Attrition")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)


target_corr = corr['Attrition'].drop('Attrition').sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(x=target_corr.values, y=target_corr.index, palette='coolwarm')
plt.title("Feature Correlation with Attrition")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Feature")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split features and target
X = df.drop(columns=['Attrition'])  # all features
y = df['Attrition']                 # target

# Train-test split
X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=1,
    stratify=y  # for balance in train/validation
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier(random_state=1)
dt_clf.fit(X_train, y_train)
```

```python
train_preds_dt_clf = dt_clf.predict(X_train)
valid_preds_dt_clf = dt_clf.predict(X_valid)

print(
    "აკურატულობა სატრენინგო მონაცემებზე: "
    f"{accuracy_score(y_train, train_preds_dt_clf) * 100:.2f}%"
)
print(
    "აკურატულობა ვალიდაციის მონაცემებზე: "
    f"{accuracy_score(y_valid, valid_preds_dt_clf) * 100:.2f}%"
)
```

```python
rf_clf = RandomForestClassifier(random_state=1)
rf_clf.fit(X_train, y_train)
```

```python
train_preds_rf_clf = rf_clf.predict(X_train)
valid_preds_rf_clf = rf_clf.predict(X_valid)

print(
    "აკურატულობა სატრენინგო მონაცემებზე: "
    f"{accuracy_score(y_train, train_preds_rf_clf) * 100:.2f}%"
)
print(
    "აკურატულობა ვალიდაციის მონაცემებზე: "
    f"{accuracy_score(y_valid, valid_preds_rf_clf) * 100:.2f}%"
)
```
